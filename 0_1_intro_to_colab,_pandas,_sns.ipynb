```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure()
sns.histplot(wage_df, x='wage')
plt.show()
```

```python
plt.figure()
sns.histplot(wage_df, x='wage', kde=True)
plt.show()
```

```python
plt.figure()
sns.histplot(wage_df, x='wage', hue='jobclass', kde=True)
plt.show()
```

```python
plt.figure()
sns.scatterplot(wage_df, x='age', y='wage')
plt.show()
```

```python
sns.scatterplot(wage_df, x='age', y='wage', hue='education')
```

```python
sns.regplot(data=wage_df, x="age", y="wage", line_kws = {"color": "red"})
```

```python
sns.regplot(data=wage_df, x="age", y="wage", order=2, line_kws={"color": "red"})
```

```python
sns.pairplot(wage_df, vars=['year', 'age', 'wage'], hue='education')
```
